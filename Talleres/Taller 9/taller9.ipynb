```python
import numpy as np
from scipy.stats import norm, hypergeom
import matplotlib.pyplot as plt

# Hypergeometric distribution
# https://en.wikipedia.org/wiki/Hypergeometric_distribution
N = 33
n = 7
M = 4
pr = hypergeom(N, M, n).pmf(k=0)
print(1-pr)

# Normal distribution
# https://en.wikipedia.org/wiki/Normal_distribution

def estandarize(x, mu, sigma):
	return (x - mu) / sigma 

mu = 27
sigma = 14
x1 = 32
x2 = 29
pr1 = norm().cdf(estandarize(x1, mu, sigma)) # 32
pr2 = norm().cdf(estandarize(x2, mu, sigma)) # 29
print(pr1, pr2)
print(pr1-pr2)
x = np.linspace(mu-3*sigma, mu+3*sigma, 100)
plt.plot(x, norm(mu, sigma).pdf(x))
plt.vlines(mu, 0, norm(mu, sigma).pdf(mu)+0.01, color='r', linestyle='--')
plt.ylim(0, norm(mu, sigma).pdf(mu)+.001)
plt.show()
norm(mu, sigma).ppf(0.75)
```
