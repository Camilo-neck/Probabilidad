```python
x = np.array([1.4, 2.4, 4.0, 4.9, 5.7, 6.3, 7.8, 9.0, 9.3, 11.0])
y = np.array([2.3, 3.7, 5.7, 9.9, 6.9, 15.8, 15.4, 36.9, 34.6, 53.2])
# a)
plt.scatter(x, y)
np.corrcoef(x, y)[0][1]
```

a) La relación entre x y y no es lineal, tiene un forma de curva exponencial.

```python
# b)
log_y = np.log(y)
plt.scatter(x, log_y)
```

b) La relación entre x y ln(y) es lineal.
