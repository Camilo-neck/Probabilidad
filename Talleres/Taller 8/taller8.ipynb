```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
"""
Punto 1
"""
# a)
x = np.arange(7) # [0, 1, 2, 3, 4, 5, 6]
Fx = np.array([0.06, 0.19, 0.39, 0.67, 0.92, 0.97, 1])
fx = map(lambda _, i: Fx[i]-Fx[i-1] if i > 0 else Fx[i], Fx, x)
fx = np.array(list(fx))
print("fx: ", fx)
# b)
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
ax[0].bar(x, fx, width=0.05, color='b')
ax[0].scatter(x, fx, marker='s', c='b')
ax[0].set_title("fx")
ax[0].set_xlabel("x")
ax[0].set_ylabel("f(x)")

ax[1].stairs(Fx[1:], x, color='b', baseline=None)
ax[1].scatter(x[1:]-1, Fx[1:], marker='s', c='b')
ax[1].set_title("Fx")
ax[1].set_xlabel("x")
ax[1].set_ylabel("F(x)")
plt.show()

# d)
mean = sum(x*fx)
var = sum(((x-mean)**2)*fx)
std = np.sqrt(var)
cv = std/mean
print("mean: ", mean)
print("cv: ", round(cv, 4)*100 , "%")
```

```python
"""
Punto 2
"""
# c)
def fx(x):
	return (3/8)*((7-x)**2)
def Fx(x):
	return 1-((1/8)*((7-x)**3))
x = np.linspace(5, 7, 1000)
f_x = fx(x)
F_x = Fx(x)
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
ax[0].plot(x, f_x, color='b')
ax[0].set_xlabel("x")
ax[0].set_ylabel("f(x)")

ax[1].plot(x, F_x, color='r')
ax[1].set_xlabel("x")
ax[1].set_ylabel("F(x)")
```

```python
"""
Punto 3
"""
# c)
def fx(x: np.ndarray, v: float):
	return v*(x**(v-1))

x = np.linspace(0, 1, 1000)
f_x1 = fx(x, 1/3)
f_x2 = fx(x, 1/2)
f_x3 = fx(x, 1)
f_x4 = fx(x, 2)
f_x5 = fx(x, 3)

plt.plot(x, f_x1, color='b', label="θ=1/3")
plt.plot(x, f_x2, color='r', label="θ=1/2")
plt.plot(x, f_x3, color='g', label="θ=1")
plt.plot(x, f_x4, color='y', label="θ=2")
plt.plot(x, f_x5, color='k', label="θ=3")

plt.title("Probabilistic density functions")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.show()
```

```python
x = np.array([200, 250, 300]) 
fx = np.array([0.3, 0.2, 0.5])

mean = sum(x*fx) # Promedio

# Método 1 : ∑(x*f(x))
var_1 = sum(((x-mean)**2)*fx)
std_1 = np.sqrt(var_1)
print("std_1: ", std_1)

# Método 2: E(x^2) - E(x)^2
mean_square_x = sum((x**2)*fx)
var_2 = mean_square_x - mean**2
std_2 = np.sqrt(var_2)
print("std_2: ", std_2)
```
