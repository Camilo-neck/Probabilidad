$$
\text{Formulas Probabilidad}
$$

**Principio multiplicativo del conteo**
$$
\begin{align*}
	n &= \prod_{i=1}^{n} p_i &= n_1 \times \dots \times n_m.
\end{align*}
$$

**Permutación sin repeticiones**	(**Permutación**)

*El orden importa y no hay repeticiones*
$$
\begin{align*}
	P_{k}^n	&= n \times (n-1) \times \dots \times (n-(k-1)) &= \frac{n!}{(n-k)!}\\
	P_{k}^n	&= \frac{n!}{(n-k)!} \\
	P_{k}^n &= nPk
\end{align*}
$$

**Permutación con repeticiones**

*El orden si importa y hay repeticiones*
$$
\begin{align*}
	n^k &= \prod_{i=1}^{k} n_i &= n \times n \times \dots \times n\\
\end{align*}
$$

**Combinación sin repeticiones** (**Combinación**)

*El orden no importa y no hay repeticiones*
$$
\begin{align*}
	C_{k}^n &= \frac{n!}{k!(n-k)!}\\
	C_{k}^n &= \binom{n}{k} = nCk
\end{align*}
$$

**Teorema aditivo**
$$
\begin{align*}
	P\left(\bigcup_{i=1}^{m} A_i\right)&=\sum_{i=1}^{m}P(A_i)-\sum_{i<j}P(A_i\cap A_j)+\sum_{i<j<k}P(A_i\cap A_j\cap A_k)-\ldots-(-1)^{m-1}P\left(\bigcap_{i=1}^{m} A_i \right)\,.
\end{align*}
$$

**Probabilidad Condicional**
$$
\begin{align*}
	P(A\mid E)&=\frac{P(A\cap E)}{P(E)}
\end{align*}
$$

**Teorema multiplicativo**
$$
\begin{align*}
	P(A\cap E)&=P(A\mid E)\,P(E)\,.
\end{align*}\\
$$
Si son eventos independientes tales que $(A\mid E)=P\left(A\right)$

$$
\begin{align*}
	P(A\cap E)&=P(A)\,P(E)\\
\end{align*}
$$

**Teorema de la probabilidad total**
$$
\begin{align*}
	\bigcup_{i=1}^m E_i=\Omega
\qquad\text{y}\qquad
E_i\cap E_j=\Phi\,,\,\text{para }i\neq j\,.
\end{align*}\\
$$
Entonces la probabilidad de un evento A se puede calcular como:

$$
\begin{align*}
	P(A) &= \sum_{j=1}^{m} P(A\mid E_j)\,P(E_j)\,.
\end{align*}
$$

**Teorema de Bayes**
$$
\begin{align*}
	P(E_k\mid A)&=\frac{P(A\mid E_k)P(E_k)}{\sum_{j=1}^{m}P(A\mid E_j)\,P(E_j)}\,.
\end{align*}
$$

**Valor esperado**	(**Media**)/(**Promedio**)
$$
\begin{equation*}
\textsf{E}[X]=\mu_X=
\left\{
  \begin{array}{ll}
    \sum_{k} x_k f_X(x_k),              & \hbox{si $X$ es una v.a.d.} \\
    \int_{-\infty}^{\infty} x f_X(x)dx, & \hbox{si $X$ es una v.a.c.}
  \end{array}
\right.
\end{equation*}
$$

En general si $g:\mathbb{R} \rightarrow \mathbb{R}$ es una función , entonces se tiene que el valor esperado de $g(X)$ se define como:
$$
\begin{equation*}
\textsf{E}[g(X)]=
\left\{
  \begin{array}{ll}
    \sum_{k} g(x_k) f_X(x_k),              & \hbox{si $X$ es una v.a.d.} \\
    \int_{-\infty}^{\infty} g(x) f_X(x)dx, & \hbox{si $X$ es una v.a.c.}
  \end{array}
\right.
\end{equation*}
$$

**Varianza**
$$
\begin{equation*}
\textsf V[X] =\sigma^2_X=
\left\{
  \begin{array}{ll}
    \sum_{k}(x_k-\mu_X)^2f_X(x_k), & \hbox{si $X$ es una v.a.d.;} \\
    \int_{-\infty}^{\infty}(x-\mu_X)^2f_X(x)dx, & \hbox{si $X$ es una v.a.c..}
  \end{array}
\right.
\end{equation*}
$$

**Desviación estándar**
$$
\textsf{DE}(X)=\sigma_X=\sqrt{\sigma_X^2}\,.
$$

**Coeficiente de variación**
$$
\textsf{CV}(X) = \left| \frac{\sigma_X}{\mu_X} \right|.
$$
