**2.10 Ejemplo**
Los siguientes datos están asociados con el peso (kg) de un conjunto de materiales. Elaborar la distribución de frecuencias correspondiente.

103.1, 82.1, 106.2, 100.9, 91.8 , 96.1 , 126.9, 119.8, 93.1, 86.8, 75.2, 93.0 , 82.3 , 94.8, 64.2 , 105.3, 108.0, 86.3 , 81.8 , 138.1, 92.5, 66.3, 66.6, 142.2, 96.5 , 74.8, 95.4 , 100.1, 81.9 , 112.0, 116.8, 103.2, 66.1, 60.4, 78.7

Barras de metal

La variable “peso” es una variable cuantitativa de razón. Es claro que esta variable no está dada en categorías, por lo que es necesario elaborar las clases pertinentes como sigue: $\\$
- Se eligen $m=6$ clases dado que $\sqrt{35}$ $=5.916≈6$ y $1+3.3log10(35)=6.095≈6\\$
- $x_{min}$ $= 60.4$ y $x_{max}$ $= 142.2.\\$
- $R=$ $x_{max}$ $−$ $x_{min}$ $=142.2−60.4=81.8.\\$
- $a =$ $\frac{81.8}{6}$ $=15.63.\\$
- Las clases resultantes son: $\\$
	1. $C_{1}$ $=$ { $x$ : $l_{0}$ $\le$ x $\le$ $l_{1}$} donde $l_{0}$ $= 60.40$ y $l_{1}$ $= 60.40 + 13.63 = 74.03.\\$
	2. $C_{2}$ $=$ { $x$ : $l_{1}$ $\le$ x $\le$ $l_{2}$} donde $l_{1}$ $= 74.03$ y $l_{2}$ $= 74.03 + 13.63 = 87.67.\\$
	3. $C_{3}$ $=$ { $x$ : $l_{2}$ $\le$ x $\le$ $l_{3}$} donde $l_{2}$ $= 87.67$ y $l_{3}$ $= 87.67 + 13.63 = 101.30.\\$
	4. $C_{4}$ $=$ { $x$ : $l_{3}$ $\le$ x $\le$ $l_{4}$} donde $l_{3}$ $= 101.30$ y $l_{4}$ $= 101.30 + 13.63 = 114.93.\\$
	5. $C_{5}$ $=$ { $x$ : $l_{4}$ $\le$ x $\le$ $l_{5}$} donde $l_{4}$ $= 114.93$ y $l_{5}$ $= 114.93 + 13.63 = 128.57.\\$
	6. $C_{6}$ $=$ { $x$ : $l_{5}$ $\le$ x $\le$ $l_{6}$} donde $l_{5}$ $= 128.57$ y $l_{6}$ $= 128.57 + 13.63 = 142.20.\\$

Así, la distribución de frecuencias está dada por:

| Clase             |   F. Absoluta | F. Relativa   | F. Abs. Acumulada   | F. Rel. Acumulada   |
|:------------------|--------------:|:--------------|:--------------------|:--------------------|
| $[60.40;74.03]$   |             5 | 14.3%         | 5                   | 14.3%               |
| $(74.03;87.67]$   |             9 | 25.7%         | 14                  | 40.0%               |
| $(87.67;101,30]$  |            10 | 28.6%         | 24                  | 68.6%               |
| $(101.30;114.93]$ |             6 | 17,1%         | 30                  | 85.7%               |
| $(114.93;128.57]$ |             3 | 8.6%          | 33                  | 94.3%               |
| $(128.57;142.20]$ |             2 | 5.7%          | 35                  | 100%                |
| Total             |            35 | 100%          | N.A.                | N.A.                |


```python
# Datos
peso = [103.1, 82.1, 106.2, 100.9, 91.8,  96.1,  126.9, 119.8, 93.1, 86.8, 75.2, 93.0, 
         82.3,  94.8, 64.2,  105.3, 108.0, 86.3,  81.8,  138.1, 92.5, 66.3, 66.6, 142.2,    
         96.5,  74.8, 95.4,  100.1, 81.9,  112.0, 116.8, 103.2, 66.1, 60.4, 78.7]
# Tamaño de la muestra
n = len(peso)
print(n)
```

```python
# Número de intervalos
m = round(1+3.33*np.log10(n))
print(m)
```

```python
# Rango
R = max(peso) - min(peso)
print(round(R,1))
```

```python
# Amplitud
a = R/m
print(round(a,5))
```

```python
# Limites
lim = list(map(lambda x: round((x*a)+min(peso),5), list(range(m+1))))
print(lim)
```

```python
# Frecuencias absolutas
nj = pd.cut(sorted(peso), lim, include_lowest=True).value_counts()
print(nj)
```

```python
# Frecuencias relativas
hj = nj.div(n)
print(hj)
```

```python
# Frecuencias absolutas acumuladas
Nj = nj.cumsum()
print(Nj)
```

```python
# Frecuencias relativas acumuladas
Hj = hj.cumsum()
print(Hj)
```

```python
# Tabla
tab = pd.concat([nj, hj, Nj, Hj], axis=1)
print(tab)
```

```python
# Edicion
# nombres_f <- ["Bachillerato", "Pregrado", "Maestria", "Doctorado"]
nombres_c = ["F. Absoluta", "F. Relativa", "F. Abs. Acumulada",   "F. Rel. Acumulada"]
tab.columns = nombres_c
# Imprimir tabla
print(tab.to_string())
```

```python
# Imprimir tabla a markdown (In R using ktnir)
print(tab.round(3).to_markdown())
```
